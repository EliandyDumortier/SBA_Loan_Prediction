# <span id="cleanse-id"> <span style='color: rgb(135, 135, 236)'> <span style='font-family:Calibri'> 3. Cleaning and preparing the data

### <span style="font-family:Calibri;"> 3.1 Avoiding Data Leakage </span>

- We are removing columns that could introduce data leakage.


```python
#Creating a subset with the columns I want to drop 
columns_to_drop=['ChgOffDate','DisbursementDate','BalanceGross','ChgOffPrinGr','SBA_Appv']
df.drop(columns_to_drop,axis=1,inplace=True)
df.info()
```

### <span style='color:White'> <span style='font-family:Georgia'> 3.2 Finding missin (null) values 
  

## <span style='color:Gray'> <span style='font-family:Calibri> 3.1 Finding missin (null) values'>


```python

# Calculate missing value count per column
missing_counts = df.isna().sum()

if missing_counts.sum() == 0:
    print("There are no missing values to be shown")
else:
    # Create a dataframe for visualization
    missing_df = pd.DataFrame(missing_counts, columns=["Missing Values"])
    
    # Sort by highest missing values
    missing_df = missing_df[missing_df["Missing Values"] > 0].sort_values(by="Missing Values", ascending=False)

    # Plot heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(missing_df.T, annot=True, cmap="Reds", linewidths=0.5, fmt=".0f", cbar=False)
    plt.title("Missing Values per Column", fontsize=14)
    plt.xlabel("Columns")
    plt.ylabel("")
    plt.xticks(rotation=45)
    plt.show()

    # Print missing values in descending order
    print("Missing values per column:\n", missing_df)
```

### <span style='font-family:Calibri'> 3.3 Finding duplicates
  

```python
#Finding duplicates 


num_duplicates = df.duplicated().sum() #To filter the dataset containing missing values
duplicate_rows = df[df.duplicated()]
print("The number of duplicate rows is equal to : {} \n\n".format(num_duplicates))
print("")


#Dropping the duplicates 
df.drop_duplicates(inplace=True)

print("Our data set now has {}: number of rows".format(df.shape[0]))
```

### <span style='font-family:Calibri'> 3.3 Transforming DateTime Variables
  

```python
df['ApprovalDate'] = pd.to_datetime(df['ApprovalDate'], errors='coerce')
df['ApprovalFY'] = pd.to_datetime(df['ApprovalFY'], format='%Y', errors='coerce')
df['ApprovalFY'] = df['ApprovalFY'].dt.year  # As it's just a fiscal year
```

### <span style='font-family:Calibri'> 3.5 Transforming Numerical Variables

```python
#Changing the type of money type of values : 

variables_to_transform=['DisbursementGross','GrAppv']

# Clean categorical columns using .strip()
for col in variables_to_transform:
    df[col] = df[col].str.replace('[$,]', '', regex=True).astype(float)

# Checking the type is now correct
for col in variables_to_transform:
    print(f"The new type for {col} is : {df[col].dtypes}")

```

### <span style='font-family:Calibri'> 3.5 Transforming Categorical Variables
  

```python

# Automatically detect categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

# Clean categorical columns in a memory-efficient way
df[categorical_columns] = df[categorical_columns].astype(str).apply(lambda x: x.str.strip())

# Set a limit for the number of categories to display
max_categories = 20  # Adjust this as needed

# Plot unique values and their counts for each categorical column
for col in categorical_columns:
    value_counts = df[col].value_counts()  # Compute once for efficiency
    top_values = value_counts[:max_categories]  # Show only top N categories

    plt.figure(figsize=(8, 5))
    bars = plt.barh(top_values.index, top_values.values, color=plt.cm.viridis_r(range(len(top_values))))
    plt.xlabel("Count", fontsize=12)
    plt.ylabel(col, fontsize=12)
    plt.title(f"Distribution of {col} (Top {max_categories})", fontsize=14)
    plt.gca().invert_yaxis()  # Invert for better readability

    # Add count labels to each bar
    for bar in bars:
        plt.text(bar.get_width() + 500, bar.get_y() + bar.get_height()/2, 
                 f'{int(bar.get_width())}', ha='left', va='center', fontsize=10)

    plt.show()
```
