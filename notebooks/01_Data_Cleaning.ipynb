# <span id="cleanse-id"> <span style='color: rgb(135, 135, 236)'> <span style='font-family:Calibri'> 3. Cleaning and preparing the data

### <span style='color:White'> <span style='font-family:Georgia'> 3.1 Finding missin (null) values 


```python

# Calculate missing value count per column
missing_counts = df.isna().sum()

if missing_counts.sum() == 0:
    print("There are no missing values to be shown")
else:
    # Create a dataframe for visualization
    missing_df = pd.DataFrame(missing_counts, columns=["Missing Values"])
    
    # Sort by highest missing values
    missing_df = missing_df[missing_df["Missing Values"] > 0].sort_values(by="Missing Values", ascending=False)

    # Plot heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(missing_df.T, annot=True, cmap="Reds", linewidths=0.5, fmt=".0f", cbar=False)
    plt.title("Missing Values per Column", fontsize=14)
    plt.xlabel("Columns")
    plt.ylabel("")
    plt.xticks(rotation=45)
    plt.show()

    # Print missing values in descending order
    print("Missing values per column:\n", missing_df)
```

- #### <span style='color:White'> <span style='font-family:Georgia'> As we found null values in the dependent variable and that we do not want to get rid of the rows containing it, and seeing that we have the 'ChgOffDate' Column we assume that this column will have a date when there was a 'ChgOff' event, to confirm this theory we are going to see the coefficiente of relationship between the two variables

```python

#Creating a copy of my DF to not do these modification in my final DATASET 
df_copy=df.copy()
# Step 1: Create a copy of the DataFrame


# Step 2: Create the 'ChgOffDate' column
df_copy['ChgOffDate'] = df_copy['ChgOffDate'].apply(lambda x: 1 if pd.notna(x) else 0)

# Step 3: Replace 'MIS_Status' values (assuming 'ChgOff' corresponds to 1 and 'PIF' to 0)
df_copy['MIS_Status'] = df_copy['MIS_Status'].apply(lambda x: 1 if x == 'CHGOFF' else 0)

# Step 4: Calculate the correlation between 'ChgOffDate' and 'MIS_Status'
correlation = df_copy['ChgOffDate'].corr(df_copy['MIS_Status'])

# Output the correlation result
print(f"Correlation between 'ChgOffDate' and 'MIS_Status': {correlation}")
```

```python
#After confirming our doubts, we are going to  Fill missing MIS_Status values with 'ChgOff' if ChgOffDate is not null else we'all assumme is PIF
df['MIS_Status'] = np.where(df['MIS_Status'].isna() & df['ChgOffDate'].notna(), 'CHGOFF', df['MIS_Status'])
df['MIS_Status'] = np.where(df['MIS_Status'].isna(), 'P I F', df['MIS_Status'])

# Calculate missing value count per column
missing_counts = df.isna().sum()
    # Create a dataframe for visualization
missing_df = pd.DataFrame(missing_counts, columns=["Missing Values"])
    
    # Sort by highest missing values
missing_df = missing_df[missing_df["Missing Values"] > 0].sort_values(by="Missing Values", ascending=False)
print("Missing values per column:\n", missing_df)
```

### <span style='font-family:Calibri'> 3.3 Transforming DateTime Variables
  

```python
#Verifying if the fiscal year is in the correct format (as per the waring of column 9 when importing)

df['ApprovalFY'].unique()
```

```python
# Remove the 'A' from the 'ApprovalFY' column (if it exists) and convert the column to numeric
df['ApprovalFY'] = pd.to_numeric(df['ApprovalFY'].astype(str).str.replace(r'[^\d]', '', regex=True), errors='coerce')
print(df['ApprovalFY'].dtypes)
```

```python
#Fixing approval_date

def adjust_year(date):
    # Split the date into day, month, and year
    day, month, year = date.split('-')
    
    # If the year is less than 50, assume it's 2000s; otherwise, it's 1900s
    if int(year) < 50:
        year = '20' + year
    else:
        year = '19' + year
        
    return f"{day}-{month}-{year}"

# Apply the adjustment to your 'ApprovalDate' column
df['ApprovalDate'] = df['ApprovalDate'].apply(adjust_year)

# Convert to datetime
df['ApprovalDate'] = pd.to_datetime(df['ApprovalDate'], format='%d-%b-%Y')
print(df['ApprovalDate'].dtype)  # Should print datetime64[ns]
```

```python
#Transforming the approvaldate variable

df['ApprovalMonth'] = df['ApprovalDate'].dt.month.astype(str)
df.drop(columns=['ApprovalDate'], inplace=True)
```

### <span style='font-family:Calibri'> 3.5 Transforming Categorical Variables
  

```python

# Automatically detect categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

# Clean categorical columns in a memory-efficient way
#df[categorical_columns] = df[categorical_columns].astype(str).apply(lambda x: x.str.strip())
df[categorical_columns] = df[categorical_columns].apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Set a limit for the number of categories to display
max_categories = 50  # Adjust this as needed

# Plot unique values and their counts for each categorical column
for col in categorical_columns:
    value_counts = df[col].value_counts()  # Compute once for efficiency
    top_values = value_counts[:max_categories]  # Show only top N categories

    plt.figure(figsize=(12, 10))
    bars = plt.barh(top_values.index, top_values.values, color=plt.cm.viridis_r(range(len(top_values))))
    plt.xlabel("Count", fontsize=12)
    plt.ylabel(col, fontsize=12)
    plt.suptitle(f"Distribution of {col} (Top {max_categories})", fontsize=14)
    plt.title(f"Number of modalities: {df[col].nunique()}", fontsize=12)
    plt.gca().invert_yaxis()  # Invert for better readability

    # Add count labels to each bar
    for bar in bars:
        plt.text(bar.get_width() + 500, bar.get_y() + bar.get_height()/2, 
                 f'{int(bar.get_width())}', ha='left', va='center', fontsize=10)

    plt.show()
```

```python
import pgeocode

df['Zip'] = df['Zip'].astype(str).str.zfill(5)  # Ensures all ZIPs have 5 digits

# Initialize the geocoder for the US
nomi = pgeocode.Nominatim('US')

# Function to fetch city and state from ZIP code
def get_city_state(zipcode):
    location = nomi.query_postal_code(zipcode)
    if pd.notna(location.place_name):
        return pd.Series([location.place_name, location.state_code])
    else:
        return pd.Series([np.nan, np.nan])  # Keep NaNs consistent

# Apply the function only for missing values
for idx, row in df.loc[df['City'].isna() | df['State'].isna()].iterrows():
    city, state = get_city_state(row['Zip'])
    
    # Update only if values are found
    if pd.notna(city):
        df.at[idx, 'City'] = city
    if pd.notna(state):
        df.at[idx, 'State'] = state

#dropping the cities and state non treated
#df = df.dropna(subset=['City', 'State'])
```
